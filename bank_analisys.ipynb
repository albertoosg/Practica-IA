#### Import the data from the simulation.csv and analyze the amount of deposits vs withdrawals made by the clients in each branch, as shown in the graph below.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import sys
import os

def analyze_income():
    df = pd.read_csv(os.path.join(os.path.dirname(os.getcwd()), "data_analisys", "simulation.csv"))
    print(df.head())
    
    # Deposits vs Withdrawals by branch.
    branch_op_type = df.pivot_table(
        values='Quantity',
        index='Branch',
        columns='Operation',
        aggfunc='sum'
    )
    branch_op_type.plot(kind='bar', figsize=(10, 6))
    plt.title('Deposits vs Withdrawals by Branch')
    plt.xlabel('Branch')
    plt.ylabel('Amount')
    plt.legend(title='Operation Type')
    plt.tight_layout()
    plt.show()
    
    return df

df = analyze_income()
```

### Cartesian Product: Clients x Operations

#### It is an analysis that generates the possible combinations between customers and types of operations. The graph below shows how many times each client has performed operations. It then indicates which operation have not been performed. 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import sys
import os
from itertools import product

def analyze_income():
    df = pd.read_csv(os.path.join(os.path.dirname(os.getcwd()), "data_analisys", "simulation.csv"))
    print(df.head())
    
    deposits = df[df['Operation'] == 'Deposit']
    
    deposits_clients = deposits.groupby('CLients')['Quantity'].sum()
    deposits_clients.sort_values(ascending=False).plot(kind = 'bar', title='Deposits by Client')
    plt.xlabel('Client')
    plt.ylabel('Total Amount')
    plt.tight_layout()
    plt.show()
    
    clients = df['CLients'].unique()
    operations = df['Operation'].unique()
    
    combinations = pd.DataFrame(product(clients, operations), columns=['Client', 'Operation'])
    
    combinations['Done'] = combinations.apply(lambda row: ((df['CLients'] == row['Client']) & (df['Operation'] == row['Operation'])).any(), axis=1)
    
    remaining = combinations[~combinations['Done']]
    print("Remaining operations by client:")
    print(remaining)
    
    table = pd.crosstab(df['CLients'], df['Operation'])
    print("Summary of clients operations: ")
    print(table)
    
    return df

df = analyze_income()
```
